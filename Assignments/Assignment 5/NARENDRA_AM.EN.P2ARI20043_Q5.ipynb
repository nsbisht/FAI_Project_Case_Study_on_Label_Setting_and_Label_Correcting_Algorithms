**Write a program that uses using Backtracking Search to solve the problem of placing k knights on an n√ón chessboard such that no two knights are attacking each other.**

**The program should either return a valid placement, or FALSE if no such placement is possible (because k is too large in relation to n).**

```python
from copy import deepcopy
```

```python
def create_empty_chessboard(n):
    board = [[ [] for _ in range(n)] for _ in range(n)]
    for x in range(n):
        for y in range(n):
            board[x][y] = '_'
    return board
```

```python
def display_chessboard(board):
    n = len(board)
    for x in range(n):
        for y in range(n):
            print(board[x][y], end = " ")
        print()
    print()
```

```python
def mark_attacking_positions(board, position):

    n = len(board)
    i = position[0]
    j = position[1]
    
    if (0 <= (i - 2) <= (n - 1)) and (0 <= (j + 1) <= (n - 1)):
        board[i - 2][j + 1] = 'A'
        
    if (0 <= (i - 1) <= (n - 1)) and (0 <= (j + 2) <= (n - 1)):
        board[i - 1][j + 2] = 'A'
        
    if (0 <= (i + 1) <= (n - 1)) and (0 <= (j + 2) <= (n - 1)):
        board[i + 1][j + 2] = 'A'
        
    if (0 <= (i + 2) <= (n - 1)) and (0 <= (j + 1) <= (n - 1)):
        board[i + 2][j + 1] = 'A'
        
    if (0 <= (i + 2) <= (n - 1)) and (0 <= (j - 1) <= (n - 1)):
        board[i + 2][j - 1] = 'A'
        
    if (0 <= (i + 1) <= (n - 1)) and (0 <= (j - 2) <= (n - 1)):
        board[i + 1][j - 2] = 'A'
        
    if (0 <= (i - 1) <= (n - 1)) and (0 <= (j - 2) <= (n - 1)):
        board[i - 1][j - 2] = 'A'
        
    if (0 <= (i - 2) <= (n - 1)) and (0 <= (j - 1) <= (n - 1)):
        board[i - 2][j - 1] = 'A'
```

```python
def is_unoccupied(board, position):
    i, j = position
    if board[i][j] == '_':
        return True
    return False
```

```python
def place_knight(board, position):
    i, j = position
    board[i][j] = 'K';
    mark_attacking_positions(board, position)
```

```python
def place_knights(number_of_knights, start_position, board):
    
    global count
    n = len(board)
    x, y = start_position
    
    if number_of_knights == 0:
        count = count + 1
        print(f"Placement {count}")
        display_chessboard(board)
        print()
        return
    
    for i in range(x, n):
        for j in range(y, n):
            if is_unoccupied(board, (i,j)):
                new_board = deepcopy(board)
                place_knight(new_board, (i,j))
                place_knights(number_of_knights - 1, (i,j), new_board)
        y = 0
```

```python
def solve(n, k):
    global count
    count = 0
    board = create_empty_chessboard(n)
    if (0 < n) and (0 < k <= n**2):
        place_knights(k, (0, 0), board) 
    if count == 0:
        print("FALSE")
    print(f"Number of possible placements = {count}")
```
